## Business Problem:
Netflix is encountering intensified competition from emerging and established streaming services like Disney+, HBO Max, Amazon Prime Video, and others. As the streaming landscape becomes increasingly crowded, it is crucial for Netflix to differentiate itself by gaining a deeper understanding of its users' behavior, preferences, and viewing habits.

To stay ahead of the competition, Netflix needs to leverage data-driven insights to:

Enhance Content Strategy: Create and acquire content that aligns more closely with the tastes of its diverse user base, fostering greater engagement and satisfaction.

Improve Personalization: Deliver more personalized content recommendations by analyzing user behavior patterns, which increases user retention and time spent on the platform.

Optimize Marketing Campaigns: Tailor marketing strategies by targeting user segments based on their viewing preferences, genres, and engagement patterns, making campaigns more efficient and cost-effective.

By understanding the underlying trends in user behavior, Netflix can refine its offerings, improve user experience, and ultimately maintain a competitive edge in a rapidly evolving streaming market.

7) No Data Privacy Issues: The data used is anonymized, ensuring user privacy is maintained while still providing valuable insights into behavior patterns.

8) Content Consumption Completeness: The dataset includes all relevant details about the content, such as show titles, genres, and viewing times, allowing for detailed content preference analysis.

9) Real-Time Data: The dataset reflects up-to-date user behavior, capturing the most recent viewing trends and habits for analysis.

10) Behavioral Consistency: Users’ behavioral patterns in the dataset are assumed to be consistent with their typical usage, with no major deviations or one-time anomalies (e.g., accidental viewing spikes).

#  Visualizing Binge Behavior

```python
# Create a binary column for binge-watching (let's assume binge-watching is defined as watching 3 or more episodes in a day)
df['Binge-Watching'] = np.where(df['Watch Time (hours)'] >= 3, 1, 0)

# Visualize binge-watching behavior by genre
binge_by_genre = df.groupby(['Genre', 'Binge-Watching']).size().unstack().fillna(0)
binge_by_genre.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm')
plt.title('Binge-Watching Behavior by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.legend(title="Binge-Watching", labels=['Not Binge-Watching', 'Binge-Watching'])
plt.show()
```
